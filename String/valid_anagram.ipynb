Given two strings s and t, return true if t is an anagram of s, and false otherwise.

 

Example 1:

Input: s = "anagram", t = "nagaram"

Output: true

Example 2:

Input: s = "rat", t = "car"

Output: false

 

Constraints:

1 <= s.length, t.length <= 5 * 104

s and t consist of lowercase English letters.

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) == len(t):
            s_map, t_map = {}, {}
            for i in range(len(s)):
                s1,t1 = s[i], t[i]
                s_map[s1] = 1 + s_map.get(s1,0)
                t_map[t1] = 1 + t_map.get(t1,0)

            for i in s_map:
                if i not in t_map:
                    return False
                else:
                    if s_map[i] != t_map[i]:
                        return False
            return True 
        else:
            return False
```
